```python
!pip install transformers
import torch
from transformers import pipeline

generate_text = pipeline(model = "databricks/dolly-v2-3b", torch_dtype = torch.bfloat16, trust_remote_code = True, device_map = "auto")
```

```python
generate_text("what is a cell?")
```

```python
generate_text("what is used to cure low haemoglobin?")
```

```python
generate_text("what is paracetamol?")
```

```python
generate_text("what is plasma?")
```
